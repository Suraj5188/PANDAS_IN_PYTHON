# **Pandas Groupby**

The GROUP BY statement groups rows that have the same values into summary rows
The groupby is one of the most frequently used Pandas functions in data analysis. It is used for grouping the data points (i.e. rows) based on the distinct values in the given column or columns. We can then calculate aggregated values for the generated groups.

```python
import pandas as pd 
import numpy as np
data={"Name":["Suraj","om","Aashu"],
      "Age":[21,16,22]
      }
df=pd.DataFrame(data)
print(df)
```

# **Creating a DataFrame with this dataset**

```python
import pandas as pd
import numpy as np

df1=pd.read_csv("/content/drive/MyDrive/Demo_DATA_csv_files/matches.csv")
print(df1)
```

# *Single aggregation*

```python
#we will calculate average win_by_runs team
df1.groupby('winner')['win_by_runs'].mean()
```

# *Multiple aggregations*

```python
#we will calculate the win_by_runs, win_by_wickets by winner team
df1.groupby('winner')[['win_by_runs','win_by_wickets']].mean()
```

# **Multiple aggregations — 2**

We can also use the agg function for calculating multiple aggregate values

```python
df1.groupby('winner')['win_by_runs'].agg(['mean','max'])
```

# **Named aggregations**

In the previous two examples, it’s not clear what the aggregate columns represent. For instance, “mean” does not tell us it is the mean of stock quantity. In such cases, we can use named aggregations.

```python
df1.groupby('winner').agg(
    avg_win_by_runs=('win_by_runs','mean'),
    max_win_by_runs=('win_by_runs','max')
)
```

# *Named aggregations using different columns*

```python
#Named aggregations using different columns
df1.groupby('winner').agg(
    avg_win_by_runs=('win_by_runs','mean'),
    avg_win_by_wickets=('win_by_wickets','mean')
)
```

# Multiple aggregations and multiple functions

```python
df1.groupby('winner')[['win_by_runs','win_by_wickets']].agg(['mean','max'])
```

# **as_index parameter**

If the output of a groupby operation is DataFrame, the group values are shown in the index. We can make them a column in the DataFrame using the as_index parameter. 

```python
df1.groupby('winner',as_index=False).agg(
    avg_win_by_runs=('win_by_runs','mean'),
    max_win_by_runs=('win_by_runs','max')
)
```
